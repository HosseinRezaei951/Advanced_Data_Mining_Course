```python
import pandas as pd

dataPath = "drive/MyDrive/Data Mining/DM_data.xlsx"
df = pd.read_excel(dataPath)
df = df.drop(columns=["Unnamed: 0"])
df
```

```python
for column in df.columns:
  df[column]= df[column].astype("float64")

print("\nData type of each column of Dataframe(after convertion): \n", df.dtypes)
```

```python
import numpy as np

nullRows = np.array([list(pd.isnull(df[column])) for column in df.columns])  
isNull = False
for i in range(len(nullRows)):
  if True in nullRows[i]:
    print("There is some null cells in " + str(i+2) + " column.")
    isNull = True

if isNull == False:
  print("There is'nt any null cell every rows in data.")
```

```python
feature_list = [column for column in df.columns if column != "deltaG"]
features = df[feature_list]
labels = df["deltaG"]
```

```python
from sklearn.model_selection import train_test_split

train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size=0.30)
print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)
```

```python
# Import the model we are using
from sklearn.ensemble import RandomForestRegressor

# Instantiate model with 1000 decision trees
rf = RandomForestRegressor(n_estimators = 1000, n_jobs=10)

# Train the model on training data
rf.fit(train_features, train_labels)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

# Use the forest's predict method on the test data
predictions = rf.predict(test_features)

# Print out the mean absolute error (mae)
print('Mean Absolute Error:', round(mean_absolute_error(predictions, test_labels), 2), 'degrees.')

# Print out the mean squared error (mse)
print('Mean Squared Error:', round(mean_squared_error(predictions, test_labels), 2), 'degrees.')
```

```python
from sklearn.metrics import r2_score

# Calculate the absolute errors
errors = abs(predictions - test_labels)

# Calculate mean absolute percentage error (MAPE)
mape = abs(100 * (errors / test_labels))

# Calculate and display accuracy
accuracy = 100 - np.mean(mape)

print('Accuracy:', round(accuracy, 2), '%.')
```
